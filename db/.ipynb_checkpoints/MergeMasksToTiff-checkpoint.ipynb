```python
stack_sums = []
for mask in mask_names[:]:
    stack = [slice_ for slice_ in masks if "%s_" % (mask) in slice_]
    stack_imgs = []
    for s in stack:
        img = Image.open("MasksFilled/%s" % s).convert('RGB')
        im_arr = np.asarray(img, dtype=int)
        im_arr = im_arr.sum(axis=2)
        im_arr[np.nonzero(im_arr)] = 1
        stack_imgs.append(im_arr)
    print mask
    print np.sum(stack_imgs)
    stack_sums.append(np.sum(stack_imgs))
```

```python
# testing out ordering the mask names by the size of the stack
order, mask_names_ordered = (list(t) for t in zip(*sorted(zip(stack_sums, mask_names))))
order = list(reversed(order))
mask_names_ordered = list(reversed(mask_names_ordered))
print mask_names_ordered[:10]
pickle.dump(mask_names_ordered, open("mask_order.p", "wb"))
```
