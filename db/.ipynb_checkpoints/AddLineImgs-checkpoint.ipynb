```python
# export jpegs from ETL (Martin Haesemeyer) tiffs

tiffs = sorted([ f for f in os.listdir('./ETL') if not f.startswith('.' ) ])
for tif in tiffs[:]:
    stack_number = tif.split('_')[0]
    stack_name = 'MH_{}-Gal4'.format(stack_number)
    tiff_stack = Tiff.imread('./ETL/{}'.format(tif))
    
    #Tiff.imsave('./Tiffs/{}.tif'.format(stack_name), tiff_stack)
    folder_path = './Stacks/{}'.format(stack_name)

    try:
        os.makedirs(folder_path)
    except OSError as exception:
        pass
        
    mean_intensity = np.mean(tiff_stack)
    print stack_name, mean_intensity
    
    if mean_intensity < 3.5:
        white_point = 40
    else:
        white_point = 50

    for i, t in enumerate(tiff_stack):
        img_path = '{}/{}-{}.jpg'.format(folder_path, stack_name, i)
        #io.imsave(img_path, t)
        if white_point:
            io.imsave(img_path, t)
            subprocess.check_output(["convert", img_path, "-format", "jpg", "-level", "0%%, %i%%" % (white_point) , img_path], stderr=subprocess.STDOUT)
```

```python
# export images from hdf5s and create entries for them in database (old way -> use adapthist, new way -> adjust gamma)

counter = 0; # index of the first set of line images (Anatomy HDF5, Owen Randlett)
counter2 = 4002; # index of the second set of line images (ZBB HDF5, Harold Burgess)
for k, key in enumerate(f2.keys()[:]):
    #stack = f.get(key)
    #stack_name = key.split('_')[0]
    stack = f2.get(key)
    stack_name = key
    folder_path = '../app/assets/images/%s/' % (stack_name)
    rel_folder_path = 'images/%s/' % (stack_name)
    print rel_folder_path
    
    try:
        os.makedirs(folder_path)
    except OSError as exception:
        print exception
        
    for i, img in enumerate(stack):
        adapthist_adjusted = exposure.equalize_adapthist(stack[i])
        img_path = '../app/assets/images/%s/%s-%i.jpg' % (stack_name, key, i)
        rel_img_path = 'images/%s/%s-%i.jpg' % (stack_name, key, i)
        io.imsave(img_path, adapthist_adjusted)
        
        insert_command = ("INSERT INTO images (image_id, line_name, image_path) values (%i, '%s', '%s')" 
                % (counter, stack_name, rel_img_path))
        
        update_command = ("UPDATE images SET image_path='%s' WHERE image_id=%i" 
                         % (rel_img_path, counter2))
        try:
            #print update_command
            cursor.execute(update_command)
            counter2 += 1
        except Exception, e:
            print repr(e)
```
