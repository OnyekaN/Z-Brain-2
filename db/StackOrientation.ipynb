### Outputting HDF5 images in Caudo-rostral and Dorsoventral orientations

```python
import os
import h5py
import numpy as np
import tifffile as Tiff
from PIL import Image
from skimage import io
from skimage import exposure
from skimage.viewer import ImageViewer
```

```python
hdf5 = h5py.File('AnatomyLabelDatabase.hdf5')
print([k for k in hdf5.keys()])
```

```python
Elav_zyx = hdf5['Elavl3-H2BRFP_6dpf_MeanImageOf10Fish'][()]
```

```python
x_len = Elav_zyx.shape[2]
y_len = Elav_zyx.shape[1]
z_len = Elav_zyx.shape[0]
print(Elav_zyx.shape)
```

```python
Elav_xzy = np.moveaxis(Elav_zyx, -1, 0)
print(Elav_xzy.shape)
```

```python
Elav_yxz = np.moveaxis(Elav_zyx, 0, 2)
print(Elav_yxz.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
image_xzy = Elav_xzy[700]
image_yxz = Elav_yxz[300]
```

```python
plot = plt.imshow(image_xzy, cmap='gray')
plt.show()
```

```python
plot = plt.imshow(image_yxz, cmap='gray')
plt.show()
```

#### Stretch array

```python
io.imsave('Orientations/Elav_yxz300.jpg', image_yxz)
io.imsave('Orientations/Elav_xzy700.jpg', image_xzy)
```
